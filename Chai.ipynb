```python
# main.py

import time

def read_temperature():
    """Simulates reading temperature from a sensor."""
    # In a real embedded system, this would read from a hardware sensor
    # For this simulation, we'll return a fixed value for demonstration
    return 25.0 # Celsius

def control_led(temperature):
    """Simulates controlling an LED based on temperature."""
    if temperature > 28.0:
        print("Temperature is high (%.1f°C). Turning LED ON (simulated)." % temperature)
        # In a real embedded system, this would turn on an LED
    else:
        print("Temperature is normal (%.1f°C). Turning LED OFF (simulated)." % temperature)
        # In a real embedded system, this would turn off an LED

def main():
    print("Embedded System Simulation Started.")
    try:
        while True:
            current_temp = read_temperature()
            control_led(current_temp)
            time.sleep(2) # Simulate delay between readings
    except KeyboardInterrupt:
        print("\nEmbedded System Simulation Stopped.")

if __name__ == "__main__":
    main()
```
